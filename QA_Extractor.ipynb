```python
import json
import numpy as np
import pandas as pd
```

```python
# Open file

f = open('./adobe_extractor/adobe-dc-pdf-services-sdk-extract-python-samples/output/TravelProtection/premium_plus.json')
data = json.load(f)
filename = "TravelProtectionPremiumPlus"
```

```python
# Finding out sequence number of titles (identified by blue underline)
ele_positions = []
for ele in range(len(data["elements"])):
    if "attributes" in list(data["elements"][ele].keys()):
        if "SpaceAfter" in list(data["elements"][ele]['attributes'].keys()):
            if data["elements"][ele]["attributes"]['SpaceAfter']==18:
                ele_positions.append(ele)
#                 print("Question:", ele)
#                 print(data["elements"][ele]["Text"])
#                 print("Answer:")
#                 print(data["elements"][ele+1]["Text"])

# Retrieving the questions (identified by blue underline) and the answers in between 
QA_data = {"Question":[], "Answer":[], "Page":[]}

for ele in range(len(ele_positions)):
    question = str()
    answer = str()
    page_number = 0
    if ele != len(ele_positions)-1:
        start = ele_positions[ele] +1
        end = ele_positions[ele+1] 
        question_asked = ele_positions[ele]
        if "Text" in list(data["elements"][question_asked].keys()):
                question = (data["elements"][question_asked]["Text"])

        for items in (np.arange(start,end,1)):
            if "Text" in list(data["elements"][items].keys()):
                answer = (data["elements"][items]["Text"])
                page = (data["elements"][items]["Page"])
    
#     print(question)
#     print(answer)
    QA_data["Question"].append(question)
    QA_data["Answer"].append(answer)
    QA_data["Page"].append(page)
    
QA = pd.DataFrame.from_dict(QA_data)
QA.to_csv(str(filename) + ".csv")
QA
```
